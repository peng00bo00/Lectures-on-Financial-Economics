## Q9.1

绝对风险厌恶系数$R_A$、相对风险厌恶系数$R_R$、绝对审慎系数$P_A$、相对审慎系数$P_R$的计算式分别为

$$
R_A = -\frac{u'' (c)}{u' (c)}
$$

$$
R_R = -\frac{c \cdot u'' (c)}{u' (c)}
$$

$$
P_A = -\frac{u''' (c)}{u'' (c)}
$$

$$
P_R = -\frac{c \cdot u''' (c)}{u'' (c)}
$$

因此只需要计算出效用函数的前三阶导数即可。

对于$u(c) = \alpha c$，有

$$
u'(c) = \alpha, \ \ \ u''(c) = u'''(c) = 0
$$

此时$R_A = R_R = 0$，不存在$P_A$和$P_R$。

对于$u(c) = c - A c^2$，有

$$
u'(c) = 1 - 2 A c, \ \ \ u''(c) = -2 A, \ \ \ u'''(c) = 0
$$

此时

$$
R_A (c) = \frac{2A}{1 - 2 A c}, \ \ \ R_R (c) = \frac{2A c}{1 - 2 A c}, \ \ \ P_A = P_R = 0
$$

对于$u(c) = -\frac{1}{c}$，有

$$
u'(c) = \frac{1}{c^2}, \ \ \ u''(c) = -\frac{2}{c^3}, \ \ \ u'''(c) = \frac{6}{c^4}
$$

此时

$$
R_A (c) = \frac{2}{c}, \ \ \ R_R (c) = 2, \ \ \ P_A (c) = \frac{3}{c}, \ \ \ P_R (c) = 3
$$

对于$u(c) = -e^{-\alpha c}$，有

$$
u'(c) = \alpha e^{-\alpha c}, \ \ \ u''(c) = -\alpha^2 e^{-\alpha c}, \ \ \ u'''(c) = \alpha^3 e^{-\alpha c}
$$

此时

$$
R_A (c) = \alpha, \ \ \ R_R (c) = \alpha c, \ \ \ P_A (c) = \alpha, \ \ \ P_R (c) = \alpha c
$$

对于$u(c) = \ln{c}$，有

$$
u'(c) = \frac{1}{c}, \ \ \ u''(c) = -\frac{1}{c^2}, \ \ \ u'''(c) = \frac{2}{c^3}
$$

此时

$$
R_A (c) = \frac{1}{c}, \ \ \ R_R (c) = 1, \ \ \ P_A (c) = \frac{2}{c}, \ \ \ P_R (c) = 2
$$

对于$u(c) = \frac{c^{1 - \rho} - 1}{1 - \rho}$，有

$$
u'(c) = \frac{1}{c^\rho}, \ \ \ u''(c) = -\frac{\rho}{c^{\rho + 1}}, \ \ \ u'''(c) = \frac{\rho (\rho + 1)}{c^{\rho + 2}}
$$

此时

$$
R_A (c) = \frac{\rho}{c}, \ \ \ R_R (c) = \rho, \ \ \ P_A (c) = \frac{\rho + 1}{c}, \ \ \ P_R (c) = \rho + 1
$$

整理成表格如下

| $u(c)$ | $\alpha c$ | $c - A c^2$              | $-\frac{1}{c}$ | $-e^{-\alpha c}$ | $\ln{c}$      | $\frac{c^{1 - \rho} - 1}{1 - \rho}$ |
| :----: | :---------:| :-----------------------:| :-------------:| :---------------:| :------------:| :---------------------------------: |
| $R_A$  | $0$        | $\frac{2A}{1 - 2 A c}$   | $\frac{2}{c}$  | $\alpha$         | $\frac{1}{c}$ | $\frac{\rho}{c}$                    |
| $R_R$  | $0$        | $\frac{2A c}{1 - 2 A c}$ | $2$            | $\alpha c$       | $1$           | $\rho$                              |
| $P_A$  | -          | $0$                      | $\frac{3}{c}$  | $\alpha$         | $\frac{2}{c}$ | $\frac{\rho + 1}{c}$                |
| $P_R$  | -          | $0$                      | $3$            | $\alpha c$       | $2$           | $\rho + 1$                          |


## Q9.2

参考课本上的证明步骤进行证明。首先，最大化期末的期望效用的一阶条件为

$$
E[u'(\tilde{w}) (\tilde{r} - r_f) ] = 0
$$

其中$\tilde{w}$为投资者将$a^*$投资于风险资产，$w_0 - a^*$投资于无风险资产，期末获得的财富

$$
\begin{aligned}
\tilde{w} &= (1 + r_f) (w_0 - a^*) + a^* (1 + \tilde{r}) \\
&= w_0 (1 + r_f) + a^* (\tilde{r} - r_f) \\
\end{aligned}
$$

一阶条件两边对$w_0$进行求导并整理可以得到$a^*$关于$w_0$的导数为

$$
\frac{\mathrm{d} a^*}{\mathrm{d} w_0} = -\frac{(1 + r_f) E[u''(\tilde{w}) (\tilde{r} - r_f)]}{E[u''(\tilde{w}) (\tilde{r} - r_f)^2]}
$$

对于任意效用函数$u$总有$u'' < 0$，因此$\frac{\mathrm{d} a^*}{\mathrm{d} w_0}$的符号与$E[u''(\tilde{w}) (\tilde{r} - r_f)]$相同。代入绝对风险厌恶系数$R_A = -\frac{u''}{u'}$，可以得到

$$
E[u''(\tilde{w}) (\tilde{r} - r_f)] = E[-u'(\tilde{w}) R_A (\tilde{w}) (\tilde{r} - r_f)]
$$

因此我们只需要分析$E[-u'(\tilde{w}) R_A (\tilde{w}) (\tilde{r} - r_f)]$的符号即可。将期望展开为按概率加权求和，可以得到

$$
E[-u'(\tilde{w}) R_A (\tilde{w}) (\tilde{r} - r_f)] = \sum_{i=1}^n p_i [-u'(w_i)] R_A (w_i) (r_i - r_f)
$$

对于$r_i \geq r_f$的情况，此时风险资产有更高的回报率即$w_i \geq w_0 (1 + r_f)$。根据IRRA的性质$R_A' \gt 0$必然有

$$
R_A (w_i) (r_i - r_f) \geq R_A [w_0 (1 + r_f)] (r_i - r_f)
$$

对于$r_i \leq r_f$的情况，则有$w_i \leq w_0 (1 + r_f)$以及$R_A (w_i) \leq R_A [w_0 (1 + r_f)]$。同样可以得到

$$
R_A (w_i) (r_i - r_f) \geq R_A [w_0 (1 + r_f)] (r_i - r_f)
$$

因此不等式对于任意情况的$r_i$都成立。由于$u' \gt 0$，对于任何情况都有

$$
[-u'(w_i)] R_A (w_i) (r_i - r_f) \leq [-u'(w_i)] R_A [w_0 (1 + r_f)] (r_i - r_f)
$$

同时由于风险资产的回报率不是无风险利率，上式在$r_i \neq r_f$的情况下严格取不等号，即

$$
\sum_{i=1}^n p_i [-u'(w_i)] R_A (w_i) (r_i - r_f) \lt \sum_{i=1}^n p_i [-u'(w_i)] R_A [w_0 (1 + r_f)] (r_i - r_f)
$$

因此

$$
\begin{aligned}
\sum_{i=1}^n p_i [-u'(w_i)] R_A [w_0 (1 + r_f)] (r_i - r_f) &= E \{ [-u'(\tilde{w})] R_A [w_0 (1 + r_f)] (\tilde{r} - r_f) \} \\
&= R_A [w_0 (1 + r_f)] E[-u'(\tilde{w}) (\tilde{r} - r_f)] \\
&= 0
\end{aligned}
$$

$$
E[-u'(\tilde{w}) R_A (\tilde{w}) (\tilde{r} - r_f)] = \sum_{i=1}^n p_i [-u'(w_i)] R_A (w_i) (r_i - r_f) \lt 0
$$

即$\frac{\mathrm{d} a^*}{\mathrm{d} w_0}$的符号恒为负号。

## Q9.3

当$\gamma = 1$时投资者在每一期的效用函数退化为对数函数$u (c) = \ln{(c)}$，此时投资者的总效用函数可以表示为

$$
U = \ln{(c_1)} + \delta E [\ln{(c_2)}]
$$

接下来考虑储蓄能够带来的效用$E [\ln{(c_2)}]$。设投资者在无风险资产上的投资为$c_f$，在风险资产上的投资为$c_p$，则第二期的期望效用为

$$
\begin{aligned}
E [\ln{(c_2)}] &= 0.5 \times \ln{(1.05 \times c_f + 1.08 \times c_p)} + 0.5 \times \ln{(1.05 \times c_f + 1.04 \times c_p)}
\end{aligned}
$$

整理后得到整体的效用函数为

$$
U = \ln{(c_1)} + 0.45 \times \ln{(1.05 c_f + 1.08 c_p)} +  0.45 \times \ln{(1.05 c_f + 1.04 c_p)}
$$

因此可以构造约束优化问题

$$
\begin{aligned}
\max \ \ \ &U = \ln{(c_1)} + 0.45 \times \ln{(1.05 c_f + 1.08 c_p)} +  0.45 \times \ln{(1.05 c_f + 1.04 c_p)} \\
\text{s.t.} \ \ \ & c_1 + c_f + c_p = 10
\end{aligned}
$$

上述问题的解析解计算较为繁琐，这里直接使用数值求解进行处理。求解后可以得到投资者在一期消费以及无风险资产、风险资产上的投资为

$$
c_1 = 5.26, \ \ \ c_f = -161.09, \ \ \ c_p = 165.82
$$

可以发现上式中$c_f < 0$，这说明需要卖空无风险资产买入风险资产进行投资从而获得最大期望效用。

如果不允许卖空，则需要引入额外的约束

$$
\begin{aligned}
\max \ \ \ &U = \ln{(c_1)} + 0.45 \times \ln{(1.05 c_f + 1.08 c_p)} +  0.45 \times \ln{(1.05 c_f + 1.04 c_p)} \\
\text{s.t.} \ \ \ & c_1 + c_f + c_p = 10 \\
& c_1 \geq 0 \\
& c_f \geq 0 \\
& c_p \geq 0 \\
\end{aligned}
$$

可以解出这种情况下投资者的资产分配为

$$
c_1 = 5.26, \ \ \ c_f = 0, \ \ \ c_p = 4.74
$$


```python
import numpy as np
from scipy.optimize import minimize

U = lambda c : np.log(c[0]) + 0.45 * np.log(1.05 * c[1] + 1.08 * c[2]) + 0.45 * np.log(1.05 * c[1] + 1.04 * c[2])

constraint = {
    'type': 'eq',
    'fun': lambda c: np.sum(c) - 10
}

x0 = np.array([10, 10, 10]) / 3
res = minimize(lambda x : -U(x), x0, constraints=constraint)
c1, cf, cp = res.x

print(f"The value of consumption in the first period, risk-free asset investment and risky asset investment is: c1={c1:.2f}, cf={cf:.2f}, cp={cp:.2f}.")
print(f"The first-period consumption is: c1={c1:.2f}. The (expected) second-period consumption is: c2={1.05*cf + 0.5 * (1.04 + 1.08) * cp:.2f}.")
```
