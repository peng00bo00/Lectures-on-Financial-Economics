## Q13.1

(a) 根据APT的结论，资产的期望回报率可以表示为

$$
\bar{r}_i = r_f + \sum_{k=1}^n \beta_{i, k} \lambda_k = r_f + \sum_{k=1}^n \beta_{i, k} ( \bar{r}_{pk} - r_f)
$$

其中$\bar{r}_{pk}$为对第$k$个因子组合的期望回报率。因此三种资产的期望回报率为

$$
\bar{r}_1 = 0.02 + 0.5 \times (0.04 - 0.02) + 1 \times (0.04 - 0.02) = 0.05
$$

$$
\bar{r}_2 = 0.02 + 0.4 \times (0.04 - 0.02) + 0.8 \times (0.04 - 0.02) = 0.044
$$

$$
\bar{r}_3 = 0.02 + 0.8 \times (0.04 - 0.02) + 0.4 \times (0.04 - 0.02) = 0.044
$$

(b) 考虑三种资产形成的组合$p$

$$
\begin{aligned}
\tilde{r}_p &= w_1 \tilde{r}_1 + w_2 \tilde{r}_2 + w_3 \tilde{r}_3 \\
&= w_1 (\alpha_1 + 0.5 \tilde{f}_1 + \tilde{f}_2) + w_2 (\alpha_2 + 0.4 \tilde{f}_1 + 0.8 \tilde{f}_2) + w_3 (\alpha_3 + 0.8 \tilde{f}_1 + 0.4 \tilde{f}_2) \\
&= (w_1 \alpha_1 + w_2 \alpha_2 + w_3 \alpha_3) + (0.5 w_1 + 0.4 w_2 + 0.8 w_3) \tilde{f}_1 + (w_1 + 0.8 w_2 + 0.4 w_3) \tilde{f}_2
\end{aligned}
$$

将组合$p$的风险完全消除掉，可以构造线性方程组

$$
\begin{bmatrix}
0.5 & 0.4 & 0.8 \\
1 & 0.8 & 0.4 \\
1 & 1 & 1 \\
\end{bmatrix}

\begin{bmatrix}
w_1 \\ w_2 \\ w_3 \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\ 0 \\ 1 \\
\end{bmatrix}
$$

求解后可以得到三种资产的组合权重为

$$
w_1 = -4, \ \ \ w_2 = 5 \ \ \ , w_3 = 0
$$

此时资产组合$p$的回报应等于无风险利率

$$
w_1 \alpha_1 + w_2 \alpha_2 + w_3 \alpha_3 = -4 \alpha_1 + 5 \alpha_2 = r_f = 0.02
$$

接下来分别构造因子组合，使得它在对应因子的载荷为1而在其它因子上的载荷为0。以第一个因子为例，得到线性方程组

$$
\begin{bmatrix}
0.5 & 0.4 & 0.8 \\
1 & 0.8 & 0.4 \\
1 & 1 & 1 \\
\end{bmatrix}

\begin{bmatrix}
w_{p11} \\ w_{p12} \\ w_{p13} \\
\end{bmatrix}
=
\begin{bmatrix}
1 \\ 0 \\ 1 \\
\end{bmatrix}
$$

求解后得到组合系数为

$$
w_{p11} = -\frac{2}{3},\ \ \ w_{p12} = 0,\ \ \ w_{p13} = \frac{5}{3}
$$

这样得到

$$
\begin{aligned}
\tilde{r}_{p1} &= w_{p11} \tilde{r}_1 + w_{p12} \tilde{r}_2 + w_{p13} \tilde{r}_3 \\
&= -\frac{2}{3} \alpha_1 + \frac{5}{3} \alpha_3 + \tilde{f}_1
\end{aligned}
$$

两边同时取期望可以得到

$$
-\frac{2}{3} \alpha_1 + \frac{5}{3} \alpha_3 = 0.04
$$

类似地，考虑第二个因子组合

$$
\begin{bmatrix}
0.5 & 0.4 & 0.8 \\
1 & 0.8 & 0.4 \\
1 & 1 & 1 \\
\end{bmatrix}

\begin{bmatrix}
w_{p21} \\ w_{p22} \\ w_{p23} \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\ 1 \\ 1 \\
\end{bmatrix}
$$

计算组合系数为

$$
w_{p21} = -\frac{2}{3},\ \ \ w_{p22} = \frac{5}{2},\ \ \ w_{p23} = -\frac{5}{6}
$$

代入第二个因子组合有

$$
\begin{aligned}
\tilde{r}_{p2} &= w_{p21} \tilde{r}_1 + w_{p22} \tilde{r}_2 + w_{p23} \tilde{r}_3 \\
&= -\frac{2}{3} \alpha_1 + \frac{5}{2} \alpha_2 - \frac{5}{6} \alpha_3 + \tilde{f}_2
\end{aligned}
$$

两边同时取期望可以得到

$$
-\frac{2}{3} \alpha_1 + \frac{5}{2} \alpha_2 - \frac{5}{6} \alpha_3 = 0.02
$$

这样可以得到关于$\alpha_1$、$\alpha_2$和$\alpha_3$的三个方程

$$
\begin{bmatrix}
-4 & 5 & 0 \\
-\frac{2}{3} & 0 & \frac{5}{3} \\
-\frac{2}{3} & \frac{5}{2} & -\frac{5}{6} \\
\end{bmatrix}

\begin{bmatrix}
\alpha_1 \\ \alpha_2 \\ \alpha_3 \\
\end{bmatrix}
=
\begin{bmatrix}
0.02 \\ 0.04 \\ 0.02 \\
\end{bmatrix}
$$

解得

$$
\alpha_1 = 0.03, \ \ \ \alpha_2 = 0.028, \ \ \ \alpha_3 = 0.036
$$

将解出的$\alpha$代入资产定价方程并取期望可以得到三种资产的期望回报率为

$$
\bar{r}_1 = 0.03 + 0.5 \times 0 + 1 \times 0.02 = 0.05
$$

$$
\bar{r}_2 = 0.028 + 0.4 \times 0 + 0.8 \times 0.02 = 0.044
$$

$$
\bar{r}_3 = 0.036 + 0.8 \times 0 + 0.4 \times 0.02 = 0.044
$$

```python
import numpy as np

## risk-free portfoliio
print("Solving weights for risk-free portfolio...")

Ap = np.array([[0.5, 0.4, 0.8],
               [  1, 0.8, 0.4],
               [  1,   1,   1]])
bf = np.array([[0], [0], [1]])

w = np.linalg.solve(Ap, bf)
w = w.flatten()

print(f"The weights of risk-free portfolio is w1={w[0]:.2f}, w2={w[1]:.2f}, w3={w[2]:.2f}.\n")

## factor 1
print("Solving weights for factor 1...")

bp1 = np.array([[1], [0], [1]])

wp1 = np.linalg.solve(Ap, bp1)
wp1 = wp1.flatten()

print(f"The weights of factor 1 is wp11={wp1[0]:.2f}, wp12={wp1[1]:.2f}, wp13={wp1[2]:.2f}.\n")

## factor 2
print("Solving weights for factor 2...")

bp2 = np.array([[0], [1], [1]])

wp2 = np.linalg.solve(Ap, bp2)
wp2 = wp2.flatten()

print(f"The weights of factor 2 is wp21={wp2[0]:.2f}, wp22={wp2[1]:.2f}, wp23={wp2[2]:.2f}.\n")

## solve allpha
print("Solving Alpha...")
Aalpha = np.array([w, wp1, wp2])
balpha = np.array([[0.02], [0.04], [0.02]])

alpha = np.linalg.solve(Aalpha, balpha)
alpha = alpha.flatten()

print(f"Alpha1 = {alpha[0]:.3f}, Alpha2 = {alpha[1]:.3f}, Alpha3 = {alpha[2]:.3f}")
```
