## Q14.1

行权价为$K - \varepsilon$的欧式买入期权多头的回报为

$$
P_1(S_T) = 
\begin{cases}
0, & S_T \leq K - \varepsilon \\
S_T - (K - \varepsilon), & S_T \gt K - \varepsilon
\end{cases}
$$

行权价为$K + \varepsilon$的欧式买入期权多头的回报为

$$
P_2(S_T) = 
\begin{cases}
0, & S_T \leq K + \varepsilon \\
S_T - (K + \varepsilon), & S_T \gt K + \varepsilon
\end{cases}
$$

行权价为$K$的欧式买入期权空头的回报为

$$
P_3(S_T) = 
\begin{cases}
0, & S_T \leq K \\
K - S_T, & S_T \gt K
\end{cases}
$$

利用上面三种期权可以构造出蝶式差价

$$
P(S_T) = P_1(S_T) + P_2(S_T) + 2 P_3(S_T)
$$

接下来开始分情况讨论。当$S_T \leq K - \varepsilon$时有

$$
P(S_T) = 0
$$

当$K - \varepsilon \lt S_T \leq K$时有

$$
P(S_T) = P_1(S_T) = S_T - (K - \varepsilon)
$$

当$K \lt S_T \leq K + \varepsilon$时有

$$
\begin{aligned}
P(S_T) &= P_1(S_T) + 2 P_3(S_T) = S_T - (K - \varepsilon) + 2 (K - S_T) \\
&= K + \varepsilon - S_T
\end{aligned}
$$

当$K \gt K + \varepsilon$时有

$$
\begin{aligned}
P(S_T) &= P_1(S_T) + P_2(S_T) + 2 P_3(S_T) \\
&= S_T - (K - \varepsilon) + S_T - (K + \varepsilon) + 2 (K - S_T) \\
&= 0
\end{aligned}
$$

整理后得到

$$
P(S_T) = 
\begin{cases}
0, & S_T \leq K - \varepsilon \\
S_T - (K - \varepsilon), & K - \varepsilon \lt S_T \leq K \\
K + \varepsilon - S_T, & K \lt S_T \leq K + \varepsilon \\
0, & K \gt K + \varepsilon
\end{cases}
$$

## Q14.2

本题目中首先算出一些常数。0时刻股票价格为$S_0 = 100$，1时刻股票价格可能为$u S_0 = 200$，也可能为$d S_0 = 50$，因此股票价格上升和下降的系数分别为

$$
u = 2.0, \ \ \ d = 0.5
$$

无风险债券在0时刻的价格为80，1时刻的价格为100，可以算出无风险利率为

$$
e^{r} = \frac{100}{80} = 1.25
$$

行权价格$K = 100$的欧式卖出期权(put)在股票上升和下降两种情况下的价格为

$$
C_u = \max \{ 110 - 200 , 0 \} = 0, \ \ \ C_d = \max \{ 110 - 50 , 0 \} = 60
$$

(1) 使用风险消除法进行计算。首先构造组合包含1只衍生品和$-\Delta$只股票，在1时刻这个组合两种情况下的支付分别为

$$
\Pi_u = C_u - \Delta u S_0 = -200 \Delta, \ \ \
\Pi_d = C_d - \Delta d S_0 = 60 -50 \Delta
$$

通过调整$\Delta$消除风险可以得到

$$
\Delta = -0.4
$$

代回0时刻组合价格的公式有

$$
\Pi_0 = C_0 - \Delta S_0 = C_0 + 40
$$

由于已经消除了风险，这个组合的回报率等于无风险利率。因此有

$$
\Pi_0 = e^{-r} \Pi_u = 64
$$

从而解出衍生品在0时刻的价格为

$$
C_0 = 24
$$

(2) 使用复制定价法进行计算。构造一个组合包含$\Delta$只股票和$B$只无风险债券，并要求这个组合在1时刻两个状态下的支付与衍生品相等，即

$$
\begin{cases}
\Delta u S_0 + e^r B &= C_u \\ 
\Delta d S_0 + e^r B &= C_d \\ 
\end{cases}
$$

可以解出股票和债券的数量为

$$
\Delta = -0.4, \ \ \ B = 64
$$

因此衍生品在0时刻的价格为

$$
C_0 = \Delta S_0 + B = 24
$$

(c) 使用风险中性定价进行计算。假设在风险中性世界中股票价格上涨和下降的概率分别为$q$和$1-q$，则股票价格满足方程

$$
S_0 = e^{-r} [q u S_u + (1-q) d S_d]
$$

从中可以解出概率为

$$
q = 0.5
$$

因此衍生品在0时刻的价格为

$$
C_0 = e^{-r} [q C_u + (1-q) C_d] = 24
$$
