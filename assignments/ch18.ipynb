## Q18.1

按照课本上的方法，首先写出$T$时刻股票的价格为

$$
S_T = S_t e^{(\mu - \frac{1}{2} \sigma^2) (T - t) + \sigma \int_t^T \mathrm{d} z_t}
$$

显然$S_T$服从对数正态分布，其期望为

$$
E [S_T] = S_t e^{\mu (T - t)}
$$

接下来变换到等鞅测度，此时股票价格在等鞅测度下符合鞅性

$$
\tilde{E} [S_T] = S_t e^{r (T - t)}
$$

因此，在等鞅测度下股票的价格需要满足

$$
S_T = S_t e^{( r - \frac{1}{2} \sigma^2 ) (T - t) + \sigma \int_t^T \mathrm{d} \tilde{z}_t}
$$

在等鞅测度下，$t$时刻欧式买入期权的价格为

$$
C_t = e^{-r (T - t)} \tilde{E} \big\{ \max \{ S_T - K \} , 0 \big\}
$$

根据前面的推导，$\ln S_T$服从正态分布

$$
\ln S_T \sim \Phi \bigg( \ln S_t + \bigg( r - \frac{1}{2} \sigma^2 \bigg) (T - t) , \sigma^2 (T - t) \bigg)
$$

令$\ln S_T = a + b u$，其中$u$是一个服从标准正态分布的随机变量，可以解得

$$
\begin{cases}
a &= \ln S_t + \bigg( r - \frac{1}{2} \sigma^2 \bigg) (T - t) \\
b &= \sigma \sqrt{T - t}
\end{cases}
$$

同样地，令$\ln K = a + b U$，可以解得

$$
U = \frac{\ln K - a}{b} = \frac{\ln K - \ln S_t - \bigg( r - \frac{1}{2} \sigma^2 \bigg) (T - t)}{\sigma \sqrt{T - t}}
$$

这样就可以求解等鞅测度下的期望了

$$
\begin{aligned}
\tilde{E} \big\{ \max \{ S_T - K \} , 0 \big\} &= \int_U^\infty (e^{a + b u} - K) \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2} u^2} \mathrm{d} u \\
&= e^{a + \frac{1}{2} b^2} [1 - N(U - b)] - K N(-U) \\
&= e^{r (T-t)} S_t N(b - U) - K N(-U)
\end{aligned}
$$

最后可以得到$t$时刻欧式买入期权的价格为

$$
C_t = S_t N(b - U) - e^{-r (T-t)} K N(-U)
$$
