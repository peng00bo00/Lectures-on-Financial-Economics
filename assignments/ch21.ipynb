# Q21.1

在存在信息不对称的情况下，企业在寻求外部融资时会优先选择信息强度高的融资方式。

按照课本中的说法，信息强度越高的融资方式越需要获取更多信息来评估其价值。这意味着在信息强度越高的融资方式中，企业越可能利用自己的信息优势来占外部投资者的便宜。因此外部投资者对信息强度高的融资方式会越审慎，越会要求更高的回报率。这样企业会倾向于使用信息强度低的融资方式来获取融资，以降低自己的融资成本。

# Q21.2

(a) 企业获取融资的约束式即为项目的净现值大于0，即

$$
(m + \eta) (R - r_f) \gt I
$$

其中$m$为不考虑宏观经济时项目成功的概率

$$
m \equiv \alpha p + (1 - \alpha) q
$$

不难发现，此时的约束式和课本中不考虑宏观经济情况下的差异在于概率$m$上添加了宏观经济景气度$\eta$的影响。

显然，当$\eta \gt 0$时原本无法满足融资约束的企业会从$\eta$得到增益从而满足新的约束式。这说明经济好的时候企业更容易获得外部融资。

反过来，当$\eta \lt 0$时原本能够满足融资约束的企业不再能够满足融资约束，即经济不好的时候企业更不容易获得外部融资。

(b) 项目的总收益可以拆分为企业以及宏观经济两部分。当经济好的时候，投资人可以从宏观经济中得到收益来补偿项目失败带来的损失。因此投资人会更愿意进行投资。
